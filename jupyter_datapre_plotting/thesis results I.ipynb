# pv dqn

```python
pv1 = pd.read_csv("E:\\final dqn\\solar\\dqn\\run1\\train_episode_R.txt")
pv2 = pd.read_csv("E:\\final dqn\\solar\\dqn\\run2\\train_episode_R.txt")
pv3 = pd.read_csv("E:\\final dqn\\solar\\dqn\\run3\\train_episode_R.txt")
pv4 = pd.read_csv("E:\\final dqn\\solar\\dqn\\run4\\train_episode_R.txt")
pv5 = pd.read_csv("E:\\final dqn\\solar\\dqn\\run5\\train_episode_R.txt")
```

```python
def calculate_sma(col):
    values = col.to_numpy()
    sma = []
    for i in range(0, values.size):
        array = values[:i + 1]
        mean_reward = np.mean(array[-100:])
        sma.append(mean_reward)
    return sma
```

```python
pvsma1 = calculate_sma(pv1["return"])
pvsma2 = calculate_sma(pv2["return"])
pvsma3 = calculate_sma(pv3["return"])
pvsma4 = calculate_sma(pv4["return"])
pvsma5 = calculate_sma(pv5["return"])
```

```python
df = pd.DataFrame()
df["ep"] = pv1["ep"]
df["pvsma1"] = pvsma1
df["pvsma2"] = pvsma2
df["pvsma3"] = pvsma3
df["pvsma4"] = pvsma4
df["pvsma5"] = pvsma5
```

```python
pvcol = df.loc[:,"pvsma1":"pvsma5"]
pvmean = pvcol.mean(axis=1)
pvstd = pvcol.std(axis=1)
print(pvmean)
```

```python
pvmean[3650:3700]
```

```python
sns.set_theme()
fig = plt.figure()
ax1 = fig.add_subplot()
x = df["ep"]

ax1.plot(x,pvmean,color="#FF7F0E")
ax1.fill_between(x,pvmean-pvstd,pvmean+pvstd,alpha=0.4,color="#FF7F0E")
ax1.set_xlabel("Episode")
ax1.set_ylabel('Mean episode return')
```

# impact of cvar term

```python
def compare_cvar_plot(file0,file1,file2,name0,name1,name2):
    df0 = pd.read_csv(file0)
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    
    labels = df1["month"].apply(lambda x: str(x)).tolist()
    x = np.arange(len(labels))  # the label locations

    base_cvar = df0["base"].tolist()
    opt0_cvar = df0["opt"].tolist()   
    opt1_cvar = df1["opt"].tolist()
    opt2_cvar = df2["opt"].tolist() 

    w = 0.2
    fig, ax = plt.subplots()
    ax.bar(x - 2*w, opt0_cvar, w, label = name0, align="edge",color="#D66600")
    ax.bar(x - w, opt1_cvar, w, label = name1, align="edge",color="#FF7F0E")
    ax.bar(x , opt2_cvar, w, label = name2, align="edge",color="#FFB069")
    ax.bar(x + w, base_cvar, w, label = "Without risk hedging ", align="edge", color="#7F7F7F")
   
    ax.set_xlabel('Month in 2019')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.text(-3, -10, "90%-CVaR of negative deviations", rotation=90, wrap=True, fontsize=12)
    ax.text(-2.3, 50, "in daily revenue (€)",rotation=90, wrap=True, fontsize=12,verticalalignment='center',)
    ax.legend(loc=1)
    fig.savefig('impact of cvar term.png', dpi=800, bbox_inches='tight')
```

```python
compare_cvar_plot(file0 = "E:\\final dqn\\cvar param\\factor0\\cvar.csv",
                  file1 = "E:\\final dqn\\solar\dqn\\run3\\cvar.csv", 
                  file2 = "E:\\final dqn\\cvar param\\factor1\\cvar.csv",
                  name0 = "Factor = 0",
                  name1 = "Factor = 0.5",
                  name2 = "Factor = 1")
```
