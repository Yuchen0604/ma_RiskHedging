```python
con_data = pd.read_csv('Forecasted_consumption_201506010000_201506302359.csv',sep=';')
con_data
```

```python
con_data['convert_Total'] = list(map(lambda x: float(x.replace(',','')),con_data['Total[MWh]']))
con_data['convert_Total'] 
```

```python
result_data['forecasted_consumption'] = con_data['convert_Total'].groupby(con_data.index // 4).sum()
result_data
```

```python
#pd.set_option('display.max_rows', None)
gen_data = pd.read_csv('Forecasted_generation_201506010000_201506302359.csv', sep=';') 
gen_data
```

```python
gen_data.dtypes
```

```python
gen_data['convert_wind_onshore'] = list(map(lambda x: float(x.replace(',','')),gen_data['Wind onshore[MWh]']))
gen_data['convert_pv'] = list(map(lambda x: float(x.replace(',','')),gen_data['Photovoltaics[MWh]']))
gen_data
```

```python
result_data['wind_offshore'] = gen_data['Wind offshore[MWh]'].groupby(gen_data.index // 4).sum()
result_data['wind_onshore'] = gen_data['convert_wind_onshore'].groupby(gen_data.index // 4).sum()
result_data['pv'] = gen_data['convert_pv'].groupby(gen_data.index // 4).sum()
result_data
```

```python
result_data['forecasted_renewables'] = result_data['wind_offshore'] + result_data['wind_onshore'] + result_data['pv']
result_data
```

```python
os.chdir(path1)
result_data.to_csv('result.csv',columns = ['date','time_of_day','day_price','yesterday_price','forecasted_renewables','forecasted_consumption','forecasted_generation'])
```
