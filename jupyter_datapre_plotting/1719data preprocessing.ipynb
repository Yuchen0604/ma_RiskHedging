# starting processing date

```python
#convert 12 hour clock to 24 hour clock
def convert_hour(time):
    in_time = datetime.strptime(time, "%I:%M %p")
    out_time = datetime.strftime(in_time, "%H")
    out_num = int(out_time)
    return out_num
# convert date to month and day
def convert_month(date):
    in_time = datetime.strptime(date, "%d-%b, %Y")
    out_time = datetime.strftime(in_time, "%m")
    out_num = int(out_time)
    return out_num
def convert_day(date):
    in_time = datetime.strptime(date, "%d-%b, %Y")
    out_time = datetime.strftime(in_time, "%d")
    out_num = int(out_time)
    return out_num
def convert_week(date):
    in_time = datetime.strptime(date, "%d-%b, %Y")
    out_time = datetime.strftime(in_time, "%w")    
    return out_time
```

```python
from datetime import datetime
test['hour'] = test['Time of day'].apply(lambda x: convert_hour(x))
test['month'] =test['Date'].apply(lambda x: convert_month(x))
test['day'] = test['Date'].apply(lambda x: convert_day(x))
test['weekday'] = test['Date'].apply(lambda x: convert_week(x))
test["week"] = test.index // 168 + 1
test.head(50)
```
