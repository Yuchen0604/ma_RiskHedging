```python
data = df2019.iloc[:,2:5]
data
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
names = data.columns
d = scaler.fit_transform(data)
scaled_data = pd.DataFrame(d, columns=names)
scaled_data.head()
```

```python
scaled_data.rename(columns={
                     "act_price": "norm_ap",
                     "fore_gen": "norm_fg",
                     "fore_price": "norm_fp",
                    }, inplace = True)
scaled_data
```

```python
norm = pd.concat([df2019, scaled_data], axis=1)
norm
```

```python
norm.to_csv("wind_2019_norm.csv", index = None)
```
